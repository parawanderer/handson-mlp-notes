```python
import numpy as np
from IPython.display import Math, display

def latex_matrix(A, fmt="{:.3g}"):
    rows = [
        " & ".join(fmt.format(x) for x in row)
        for row in A
    ]
    body = r" \\ ".join(rows)
    return r"\begin{bmatrix}" + body + r"\end{bmatrix}"

A = np.random.randn(3, 4)
display(Math(latex_matrix(A)))
```
